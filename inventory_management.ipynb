## 2.1 Add Product
Function: `add_product(product_name, price, quantity)`

```python

inventory = [] 
```

```python
def add_product(product_name: str, price: float, quantity: int): # 0. Adds a product to the inventory
    for product in inventory: # 1. Check if product exists in inventory running all the list
        if product["name"] == product_name: # 2. If the product already exists (matching name)
            product["quantity"] += quantity  # 3. Updates the quantity
            return f"Updated {product_name} quantity. New total: {product['quantity']}" # 4. Returns an updating message and stop the function
    # 5. If the loop finishes withouth finding the product, add new one 
    new_product = {"name": product_name, "price" : price, "quantity" : quantity}
    inventory.append(new_product)
    return f"Added new product: {product_name}" # 6. Return the success message
```

```python
#-----TEST AREA-----
add_product("skirt", 25.00, 10)
add_product("dress", 30.00, 15)
add_product("jeans", 35.00, 20)
add_product("hat", 20.00, 10)
add_product("scarf", 8.00, 20)
add_product("tote bag", 5.00, 30)
add_product("shirt", 35.00, 20)
add_product("t-shirt", 15.00, 30)
add_product("blazer", 40.00, 15)
add_product("coat", 60.00, 15)

```


## 2.2 View Inventory
Function: `view_inventory()`

```python
def view_inventory(): # 0. Created to visualize each product and its characteristics line by line, as simply printing the list has poor formatting
    print("\n--- Current Inventory Details ---")  
    for product in inventory: # 1. Loop through the list to print each item line by line
        print(f"Product name: {product['name']} | Price: €{product['price']} | Qty: {product['quantity']}")
```
